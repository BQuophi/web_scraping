```python
from bs4 import BeautifulSoup as soup
from urllib.request import Request, urlopen
```

```python
url = 'https://www.jumia.com.gh/phones-tablets/'
request = Request(url, headers = {'User-Agent': 'Mozilla/5.0'})
```

```python
page = urlopen(request).read()
page
```

```python
jumia_page = soup(page, 'html.parser')
jumia_page
```

```python
jumia_page.select(".info .name")
```

```python
item_names = [item.text.strip()for item in jumia_page.select('.info .name')]
item_names
```

```python
item_prices = [prices.text.strip() for prices in jumia_page.select('.info .prc')]
item_prices 
```

```python
import pandas as pd
df = pd.DataFrame({'List Of Phones and Tablets': item_names, 'Prices': item_prices})
df
```

```python
# Now we remove the spaces with no data entries
Items_table = df[:40]
Items_table 
```

```python
links_list = jumia_page.find_all('a', class_='core') 
for link in links_list:
    if 'href' in link.attrs:
        print(str(link.attrs['href']) + "\n")
```

```python
links_list = jumia_page.find_all('a', class_='core') 
for link in links_list:
    if 'href' in link.attrs:
        links = (str(link.attrs['href']) + '\n') # Same as the above but storing in a variable
        print(links)
```
